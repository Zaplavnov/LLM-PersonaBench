```python
import json
import pandas as pd
import re
import os

file_path = '../data/Palign/Test-set.json'

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Файл не найден: {file_path}")

print("Читаем и чиним JSON...")

with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()

content = content.strip()

if content.startswith('{'):
    content = '[' + content + ']'

content = re.sub(r',\s*(?=\s*[\]}])', '', content)
content = re.sub(r'("\w+"):\s*$', r'\1: null', content, flags=re.MULTILINE)

last_brace = content.rfind('}')
if last_brace != -1:
    content = content[:last_brace + 1] + content[last_brace + 1:].lstrip(', \n\r\t') 
    if content.count('{') > content.count('}'):
        content += '}' * (content.count('{') - content.count('}'))
    if content.strip().endswith('}'):
        content = content.rstrip() + ']' if not content.strip().endswith(']') else content

print(f"Длина текста после чистки: {len(content)} символов")

try:
    data = json.loads(content)
    print(f"Успешно распарсили! Загружено объектов: {len(data)}")
except json.JSONDecodeError as e:
    print(f"Ошибка на позиции {e.pos}:")
    print(content[max(0, e.pos-200):e.pos+200])
    raise

df = pd.DataFrame(data)

print(f"\nГотово! Размер датасета: {df.shape}")
print(f"Колонки: {df.shape[1]}, строки: {df.shape[0]}")

df['year'] = df['year'] + 1900
df['sex'] = df['sex'].map({1: 'Male', 2: 'Female'})

item_columns = [f'i{i}' for i in range(1, 301)]
df[item_columns] = df[item_columns].astype('Int64')

time_cols = ['sec', 'min', 'hour', 'date', 'month', 'year']

print("\nПервые 5 строк:")
display(df.head())

print("\nИнформация о датасете:")
df.info()

print("\nСтатистика по айтемам (ответы 1–5):")
display(df[item_columns].describe())


print(f"\nВсё сохранено в папку data:")
print("   → Test-set_CLEAN.csv")
print("   → Test-set_CLEAN.pkl")
```
